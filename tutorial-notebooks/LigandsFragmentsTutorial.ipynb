```python
from dinc_ensemble import load_ligand
from dinc_ensemble.ligand import DINCFragment
from glob import glob
```

## Load ligand from file

```python
possible_ligands = glob("../data/pdbbind_test_ligands/*.mol2")
```

```python
index = 0
lig_file = possible_ligands[index]
lig_file
```

```python
dinc_ligand = load_ligand(lig_file)
```

```python
dinc_ligand._rdkit_molecule
```

```python
from dinc_ensemble.ligand.core.draw import draw_flat_rdkit

draw_flat_rdkit(dinc_ligand)
```

## Fragment ligand

```python
from dinc_ensemble.parameters.fragment import *
from IPython.display import SVG

frag_params = DincFragParams(
    frag_mode=DINC_FRAGMENT_MODE.MANUAL,
    frag_size=4,
    frag_new=3, 
    root_type=DINC_ROOT_TYPE.AUTO,
    root_auto=DINC_ROOT_AUTO.FIRST,
)
frag1 = DINCFragment(dinc_ligand, frag_params)
frag1._split_to_fragments_()
```

```python
frag_df = frag1._to_df_frags_info_()
```

```python
frag_df["frag_svg"] = frag_df["frag_svg"].apply(lambda x: SVG(x))
```

```python
from dinc_ensemble.ligand.fragment.draw import draw_fragment
```

```python
draw_fragment(frag1)
```

```python
for svg in frag_df["frag_svg"]:
    display(svg)
```
