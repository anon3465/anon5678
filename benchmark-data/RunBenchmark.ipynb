```python
### Try loading all ligands
root_dir = Path("./full_dataset")
loaded_ligands = {}
error_ligands = {}
split_fragments = {}
error_fragments = {}
for i, row in full_dataset.iterrows():
    pdb_code_l = row.pdb_code.lower()
    ligand_path1 = root_dir / Path("{}".format(pdb_code_l)) / Path("{}_ligand.mol2".format(pdb_code_l))
    ligand_path2 = root_dir / Path("{}".format(pdb_code_l)) / Path("{}_peptide.mol2".format(pdb_code_l))
    ligand_path = ligand_path1
    if not ligand_path1.exists():
        ligand_path = ligand_path2
    try:
        print("~~~~~~~~~~~~~~~~~~~~~")
        print("loading ligand {} - {}".format(i, pdb_code_l))
        print("~~~~~~~~~~~~~~~~~~~~~")
        dinc_l = load_ligand(str(ligand_path))
        loaded_ligands[pdb_code_l] = dinc_l
        try:
            frag = DINCFragment(dinc_l)
            split_fragments[pdb_code_l] = frag
        except Exception as e:
            error_fragments[pdb_code_l] = e
    except Exception as e:
        error_ligands[pdb_code_l] = e

    


```

```python
times = []
pdb_codes = []
for i, row in full_dataset.iterrows():
    print(i)
    pdb_code = row.pdb_code.lower()
    start_time = time.time()
    print(pdb_code)
    lig_path = row["ligand_path"]
    rec_path = row["receptor_path"]
    cmd = " dinc-ensemble dock ../benchmark-data/{LIGAND_PATH} \
    ../benchmark-data/{RECEPTOR_PATH} \
        ../benchmark-results-try3/{PDB_CODE}_bench_test \
            --dock-type INCREMENTAL > ./benchmark_try3_outs/{PDB_CODE}_out.txt".format(LIGAND_PATH=lig_path,
                                                          RECEPTOR_PATH=rec_path,
                                                          PDB_CODE=pdb_code)
    os.system(cmd)
    
    end_time = time.time()
    pass_time = end_time-start_time
    print("TIME: {}s".format(pass_time))
    times.append(pass_time)
    pdb_codes.append(pdb_code)
df = pd.DataFrame({"pdb_code": pdb_codes, 
              "times": times})
df.to_csv("./runtimes_benchmark_t3.csv")
```

```python
    
from rdkit.Chem.Draw import rdMolDraw2D
from math import isnan
from IPython.display import SVG
from dinc_ensemble.ligand.core.convert_molecule import to_rdkit, molkit_to_rdkit

yellow = (255/255, 255/255, 0)
green = (153/255, 250/255, 157/255)
blue = (100/255, 169/255, 255/255)
red = (1, 0, 0)

def get_rdkit_bond_idx(bond_map, bond_row):
    row_bondset = set([bond_row.atom1_molkit_unique_name, 
                        bond_row.atom2_molkit_unique_name])
    for i, row in bond_map.iterrows():
        map_bondset = set([row.molkit_bonds.atom1.name, 
                           row.molkit_bonds.atom2.name])
        if map_bondset == row_bondset:
            return row["rdkit_idx"]
    return 0

def draw_fragment(frag: DINCFragment, 
                w: int = 400, h: int = 400) -> SVG:

    color_atoms = {
    }
    color_bonds = {
    }
    arads = {}
    rdkit_mol, atom_map, bond_map = molkit_to_rdkit(frag._molecule.molkit_molecule)
    frag.atoms["rdkit_idx"] = frag.atoms.apply(lambda x: \
                                               atom_map["rdkit_idx"][atom_map["name"]==x.name].iloc[0] \
                                               if atom_map["rdkit_idx"][atom_map["name"]==x.name].shape[0]>0 \
                                                else 1, \
                                                axis=1)
    frag.bonds["rdkit_idx"] = frag.bonds.apply(lambda x: get_rdkit_bond_idx(bond_map, x), axis=1) 
    frag._molecule._rdkit_molecule = rdkit_mol
    rdkit_mol = to_rdkit(frag._molecule, flatten=True)
    # color atoms in node
    for i, row in frag.atoms.iterrows():
        if isnan(row["rdkit_idx"]):
            continue
        a_id = int(row["rdkit_idx"])
        node_id = int(row["node"])
        molkit_name = row.name
        arads[a_id] = -.3
        if molkit_name == frag._root_atom_name:
            color_atoms[a_id] = [red]
        elif node_id == 0:
            color_atoms[a_id] = [yellow]
        else:
            color_atoms[a_id] = [green]

    # color active bonds between nodes
    for i, row in frag.bonds[frag.bonds.activeTors_==1].iterrows():
        if isnan(row["rdkit_idx"]):
            continue
        b_id = int(row["rdkit_idx"])
        if row["node"] != -1:
            color_bonds[b_id] = [blue]
    
    # color frozedn bonds between nodes
    for i, row in frag.bonds[(frag.bonds.activeTors_==0) & (frag.bonds.possibleTors_==1)].iterrows():
        if isnan(row["rdkit_idx"]):
            continue
        b_id = int(row["rdkit_idx"])
        if row["node"] != -1:
            color_bonds[b_id] = [yellow]

    # color bonds between atoms in the same node
    '''
    for b in rdkit_mol.GetBonds():
        b_id = b.GetIdx()
        a1_id = b.GetBeginAtomIdx()
        a1_node_id = frag.atoms[frag.atoms["rdkit_idx"]==a1_id].node.iloc[0]
        a2_id = b.GetEndAtomIdx()
        a2_node_id = frag.atoms[frag.atoms["rdkit_idx"]==a2_id].node.iloc[0]
        if a1_node_id == a2_node_id:
            color_bonds[b_id] = color_atoms[a1_id]
    '''
    print(color_atoms)
    print(color_bonds)
    d = rdMolDraw2D.MolDraw2DSVG(400, 400)
    d.DrawMoleculeWithHighlights(rdkit_mol,"",
                                dict(color_atoms),
                                dict(color_bonds),
                                arads,{})
    d.FinishDrawing()
    return SVG(d.GetDrawingText())
```
